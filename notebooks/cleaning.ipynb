```python
import pandas as pd
import numpy as np

df = pd.read_csv("heart_disease_raw.csv")
df.head()
```

```python
num_cols = ["trestbps","chol","thalch","oldpeak"]
cat_cols = ["restecg","slope"]

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
df.head()
```

```python
cat_features = ["sex","cp","fbs","restecg","exang","slope"]
for col in cat_features:
    df[col] = df[col].astype(str).str.strip().str.lower()
df.head()
```

```python
df.to_csv("heart_disease_cleaned.csv", index=False)
```
